```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python


data = {
    'hours': [1, 2, 3, 4, 5],
    'marks': [22, 31, 39, 51, 61]
}


```

```python
df = pd.DataFrame(data)
x = df['hours'].values
y = df['marks'].values

print('x:', x)
print('y:', y)
```

```python
plt.scatter (x,y)
plt.xlabel("hours studied")
plt.ylabel("marks obtained")
plt.title("hours  vs marks ")
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)

print("mean of X :",x_mean)
print("mean of y:",y_mean)
```

```python
numerator=np.sum((x-x_mean)*(y-y_mean))
denominator=np.sum((x-x_mean)**2)
a=numerator/denominator
print("slope (a) :",a)
b=y_mean - a*x_mean
print ("intercept (b) :",b)
print(f"Regression Equation : y ={a:.2f}x+{b:.2f}")
```

```python
b=y_mean - a*x_mean
print ("intercept (b) :",b)
```

```python
print(f"Regression Equation : y ={a:.2f}x+{b:.2f}")
```

```python
y_pred=a*x +b
print("Predicted Values ",y_pred)
```

```python
z=np.array([4,5,6,7])
y_pred_z=a*z+b
print("predicted values for z: ",y_pred_z)
```

```python
plt.scatter(x,y,label="actual data")
plt.plot(x,y_pred,color="red",label='regression line')
plt.xlabel("Hours studied")
plt.ylabel("marks obtained")
plt.legend()
plt.show()
```

```python
errors=y-y_pred
print("residuals:",errors)
```

```python
mse=np.mean(errors**2)
print("mean squared error:", mse)
```
