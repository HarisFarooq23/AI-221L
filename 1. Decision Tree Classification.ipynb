```python
import math

def entropy(y):
    total = len(y)
    count_0 = sum(1 for val in y if val == 0)
    count_1 = sum(1 for val in y if val == 1)
    
    p0 = count_0 / total 
    p1 = count_1 / total 
    
    ent = 0
    if p0 > 0:
        ent -= p0 * math.log2(p0)
    if p1 > 0:
        ent -= p1 * math.log2(p1)
    
    return ent

print("Parent Entropy:", entropy(y))
```

```python
X=data.iloc[:, 1:4].values
X
```

```python
def information_gain(X, y, feature_index, threshold):
    
    left_y = []
    right_y = []
    
    for i in range(len(X)):
        if X[i][feature_index] <= threshold:
            left_y.append(y[i])
        else:
            right_y.append(y[i])
    
    
    if len(left_y) == 0 or len(right_y) == 0:
        return 0
    
    total = len(y)
    
    left_entropy = entropy(left_y)
    right_entropy = entropy(right_y)
    
    weighted_entropy = (len(left_y)/total)*left_entropy + (len(right_y)/total)*right_entropy
    
    return entropy(y) - weighted_entropy
```

```python
print("IG (Age ≤ 30):", information_gain(X, y, 1, 30))
print("IG (Age ≤ 35):", information_gain(X, y, 1, 35))
print("IG (Age ≤ 44):", information_gain(X, y, 1, 44))
print("IG (Age ≤ 40):", information_gain(X, y, 1, 40))
```

```python
def build_tree(X, y, depth=0, max_depth=3):
    
    if len(set(y)) == 1 or depth == max_depth:
        return {"leaf": True, "class": max(set(y), key=y.count)}
    
    
    best_ig = -1
    best_feature = None
    best_threshold = None
    
    for feature_index in range(len(X[0])):
        values = set([row[feature_index] for row in X])
        for val in values:
            ig = information_gain(X, y, feature_index, val)
            if ig > best_ig:
                best_ig = ig
                best_feature = feature_index
                best_threshold = val
    
    
    if best_ig == 0:
        return {"leaf": True, "class": max(set(y), key=y.count)}
    
    
    left_X, left_y, right_X, right_y = [], [], [], []
    for i in range(len(X)):
        if X[i][best_feature] <= best_threshold:
            left_X.append(X[i])
            left_y.append(y[i])
        else:
            right_X.append(X[i])
            right_y.append(y[i])
    
    
    return {
        "leaf": False,
        "feature": best_feature,
        "threshold": best_threshold,
        "left": build_tree(left_X, left_y, depth+1, max_depth),
        "right": build_tree(right_X, right_y, depth+1, max_depth)
    }
```

```python
tree = build_tree(X.tolist(), y.tolist(), max_depth=3)
```

```python
import matplotlib.pyplot as plt
def plot_manual_tree(tree, feature_names, class_names, x=0.5, y=1, dx=0.2, dy=0.1, ax=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(12,8))
        ax.axis('off')
    
    if tree['leaf']:
        # Leaf node
        ax.text(x, y, f'Class: {class_names[tree["class"]]}', ha='center', bbox=dict(facecolor='lightgreen'))
        return ax
    
    # Internal node
    feature = feature_names[tree['feature']]
    threshold = tree['threshold']
    ax.text(x, y, f'{feature} <= {threshold}', ha='center', bbox=dict(facecolor='lightblue'))
    
    # Draw left child
    left_x = x - dx
    left_y = y - dy
    ax.plot([x, left_x], [y-0.01, left_y+0.01], 'k-')
    plot_manual_tree(tree['left'], feature_names, class_names, x=left_x, y=left_y, dx=dx/2, dy=dy, ax=ax)
    
    # Draw right child
    right_x = x + dx
    right_y = y - dy
    ax.plot([x, right_x], [y-0.01, right_y+0.01], 'k-')
    plot_manual_tree(tree['right'], feature_names, class_names, x=right_x, y=right_y, dx=dx/2, dy=dy, ax=ax)
    
    return ax
```

```python
feature_names = ['Gender', 'Age', 'EstimatedSalary']
class_names = ['No', 'Yes']

ax = plot_manual_tree(tree, feature_names, class_names)
plt.show()
```
